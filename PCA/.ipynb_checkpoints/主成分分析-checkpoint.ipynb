# 特征学习

### 主成分分析实现mnist数据集的降为

```python
import paddle.dataset
import numpy as np
```

```python
dataset = []
label = []
for i, data in enumerate(paddle.dataset.mnist.train()()):
    dataset.append(data[0])
    label.append(data[1])
```

```python
dataset = np.array(dataset).reshape(784, 60000)
```

```python
dataset.shape
```

### 中心化样本

```python

x = dataset

x = x - np.sum(x, axis=1, keepdims=True)/x.shape[1]
```

### 计算样本的协方差矩阵$XX^T$

```python
conv_x = np.cov(x)
```

### 求解特征值和特征向量

```python
conv_x = np.mat(conv_x)
feature_value, feature_vector = np.linalg.eig(conv_x)
```

```python
index = np.argsort(feature_value)[-2:]
```

```python
feature_vector[:2]
```

```python
w = np.array(feature_vector[index])
w = w/ np.sum(w, axis=1, keepdims=True)**2
```

```python
w = np.mat(w)
```

```python
X = w * np.mat(dataset)
```

```python
X = np.array(X.reshape(60000, 2))
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:,0], X[:,1])
```

```python
catogery = [[] for i in range(10)]
for i in range(1000):
    catogery[label[i]].append(X[i])
```

```python
for i in range(10):
    number = np.array(catogery[i])
    plt.scatter(number[:,0], number[:,1])
```
